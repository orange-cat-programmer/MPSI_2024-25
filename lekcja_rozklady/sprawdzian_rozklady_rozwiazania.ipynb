```python
#zadanie 1.
import numpy as np
from scipy.stats import gaussian_kde
from scipy.stats import norm
from matplotlib import pyplot as plt

data = np.random.binomial(n=10000, p=1/3, size=100000)
data = data - np.mean(data)
data = data/(np.var(data)**0.5)

x = np.linspace(np.min(data), np.max(data), 1000)
p = gaussian_kde(data)

fig, ax = plt.subplots(2,1)
ax[0].hist(data, bins=60, color='violet', edgecolor='b', density=True)
ax[0].set(xlabel='Values', ylabel='Probability density', title='Histogram of binomial')
ax[0].plot(x, norm.pdf(x, loc=0, scale=1))
ax[1].hist(data, bins=60, color='violet', edgecolor='b', density=True)
ax[1].plot(x, p(x))
ax[1].set(xlabel='Values', ylabel='Probability density', title='Histogram vs gaussian_kde')
fig.tight_layout()
data;

```

```python
import seaborn as sns

data = np.random.standard_cauchy(size=10000)
print(np.mean(data))
print(np.var(data))
lower = np.percentile(data, 5)
upper = np.percentile(data, 95)

data = data[(data >= lower) & (data <= upper)]

fig, ax = plt.subplots(3,1)
ax[0].hist(data, bins=50, range=(lower, upper), density=True)
sns.kdeplot(data, clip=(lower, upper), ax=ax[0], bw_adjust=0.3)

ax[1].hist(data, bins=50, range=(lower, upper), density=True)
sns.kdeplot(data, ax=ax[1], clip=(lower, upper), bw_adjust=0.5)

ax[2].hist(data, bins=50, range=(lower, upper), density=True)
sns.kdeplot(data, ax=ax[2], clip=(lower, upper), bw_adjust=1.5)
```

```python
sample_size = 100000

a = 0
for i in range(sample_size):
    sum = 0
    while sum < 21:
        roll = np.random.randint(1,7)
        if roll == 1:
            sum = 0
            break
        else:
            sum += roll
    a += sum
    
b = 0
for i in range(sample_size):
    roll = np.random.randint(1,9)
    sum = roll
    while True:
        new_roll = np.random.randint(1,9)
        if new_roll <= roll:
            break
        else:
            sum += new_roll
            roll = new_roll
    b += sum

print(a/sample_size)
print(b/sample_size)
            
        
    
```
